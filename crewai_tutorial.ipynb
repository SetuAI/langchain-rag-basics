```python
# Define our 1st Agent

from crewai import Agent

Researcher_Agent = Agent(
    role = "Market Researcher Agent",
    goal = "provides in depth information on the latest key trends in Cloud infra\
            and GPUs.s It also cites the sources from where it gets the information.\
            Avoids giving any misinformation.\
            The information is latest as of March 2026.",
            
    backstory = "An expert market research agent wth decades of experience in \
                 tech journalism. You have a unique ability to keep up with the latest trends\
                and write interesting blogs/articles around the same .",
                
    tools = [web_search_tool],
    verbose = True
            
)
```

```python
# Define the 2nd Agent

Writer_Agent = Agent(
    role = "Content Writer",
    goal = "Writes interesting and engaging blogs and articles around the latest trends\
        in the field of Cloud infra and GPUs. Has unique ability to use interesting movie\
            references to make the articles more interesting and clickable.",
    backstory = "A skilled writer with passion for understanding new age technology\
        associated market trends and researching adoption of the same.",
    
    tools = [], # no tools required here
    
    verbose = True
    
)
```

```python
# Once we are done with defining the agents,
# then we will define the tasks that shall be delegated to the agents.

from crewai import Task,Crew

# task 1 
Research_Task = Task(
    description="Search the web for the latest trends in tehe cloud infra and GPUs.\
        After that provide a summarized report, also make sure it is latest as of March 2026.",
    
    expected_output="Summary of the top 10 trending developments and insights on Cloud Infra.\
        and GPU and their impact on the businesses , especially in India and Singapore.",
        
    agent = Researcher_Agent
    
)
```

```python
# task 2 

Write_Task = Task(
    description="Write an engaging blog post on the latest \
                advancements basis the researcher agent's summary",
    expected_output="A well structured 5 paragraph long blog post or article,\
            in the markdown format with simple language being used.\
            It should have interesting references as well.",
    agent = Writer_Agent,
    output_file="blog-posts/new_post.md"
)
```

```python
# Agents are defined and so are their tasks
# now we assemble the crew

crew = Crew(
    agents = [Researcher_Agent,Writer_Agent],
    tasks = [Research_Task, Write_Task],
    verbose=False, # shows all the processes it runs , if set to True
    planning = True # you can see how are agents planning their tasks
)

# to execute
crew.kickoff()
```
